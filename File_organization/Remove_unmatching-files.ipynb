### Remove not matching files

```python
import os
```

```python
txt_dir = "/mnt/d/retraining_BirdNET/iterative_training_2025/output_results_2025/model_0/"  # Replace with the path to your .txt files directory
wav_dir = "/mnt/d/retraining_BirdNET/iterative_training_2025/input_20%/"  # Replace with the path to your .wav files directory
```

```python
def remove_unmatched_txt_files(txt_dir, wav_dir):
    """
    Remove .txt files that do not have a corresponding .wav file, considering only the shared part of their names.

    Args:
        txt_dir (str): Path to the directory containing .txt files.
        wav_dir (str): Path to the directory containing .wav files.
    """
    # Get the set of .wav filenames (without extensions)
    wav_files = {os.path.splitext(file)[0] for file in os.listdir(wav_dir) if file.lower().endswith('.wav')}

    # Process .txt files in the txt_dir
    unmatched_txt_files = []
    for txt_file in os.listdir(txt_dir):
        if txt_file.lower().endswith('.txt'):
            # Extract the shared part of the .txt filename
            txt_base_name = txt_file.split('.BirdNET.selection.table.txt')[0]
            if txt_base_name not in wav_files:
                unmatched_txt_files.append(txt_file)

    # Remove unmatched .txt files
    if unmatched_txt_files:
        print("The following .txt files do not have a corresponding .wav file and will be removed:")
        for txt_file in unmatched_txt_files:
            txt_path = os.path.join(txt_dir, txt_file)
            print(txt_path)
            os.remove(txt_path)
        print(f"Total unmatched .txt files removed: {len(unmatched_txt_files)}")
    else:
        print("All .txt files have matching .wav files.")
```

```python
# Example usage
if __name__ == "__main__":
    
    remove_unmatched_txt_files(txt_dir, wav_dir)
```
