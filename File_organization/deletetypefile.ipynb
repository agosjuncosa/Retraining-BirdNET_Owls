```python
import os
import glob

# Directory to start the search
start_directory = "/mnt/e/ARU_analisis/P14-ARU6(segundo-round)-2308-2409/grabaciones/"

# File extension to delete
target_extension = ".txt"

# Walk through directory and subdirectories
for dirpath, dirnames, filenames in os.walk(start_directory):
    # Construct the search pattern for the files
    search_pattern = os.path.join(dirpath, f"*{target_extension}")
    
    # Use glob to get all matching files
    matching_files = glob.glob(search_pattern)
    
    # Delete each matching file
    for filepath in matching_files:
        os.remove(filepath)
        print(f"Deleted: {filepath}")

print("Deletion process complete.")
```
